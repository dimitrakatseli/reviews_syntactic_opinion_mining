# Example

```python
review = nlp(u"""


Είμαι κάτοχος του τηλεφώνου εδώ και περίπου μία εβδομάδα και μπορώ να πω πως στην mid range κατηγορία τιμής δεν υπάρχει ανταγωνιστής. Η κάμερα του είναι εξαιρετική σε βίντεο και σε φωτογραφία, το ηχείο ικανοποιητικό με καλή ποιότητα και υψηλή ένταση, η gaming εμπειρία είναι τρομερή και σε γενικές γραμμές το τηλέφωνό είναι ένα τέρας. Μικρό παράπονο έχω από την μπαταρία η οποία είναι λίγο μικρή. Αν ψάχνετε μια νέα α συσκευή σε αυτό το μπάτζετ βάλτε το xiaomi mi a2 καλά στον νου σας. Είναι ένα πολύ καλό τηλεφωνο. Δεν είναι κακό 
""")
```

# Get aspects - nouns 

```python
for token in review :
    if token.pos_ =="NOUN" and token.dep_ !='obl' and token.dep_ !='obl' and token.is_stop==False:
      print(token.text, token.pos_,token.dep_)
```

# Get Opinion

```python
#get all children
for token in review :
    if token.pos_ =="NOUN":
      print(token.text, token.pos_,token.dep_,
           [child for child in token.children])
```

# Get Opinion ADJ-->NOUN 

```python
for token in review :
    if token.pos_ =="ADJ":
        for child in token.children:
            if child.pos_=='NOUN':
                print(token.text, token.pos_,token.dep_,[child])
```

# Get Opinion NOUN-->ADJ

```python
for token in review :
    if token.pos_ =="NOUN":
        for child in token.children:
            if child.pos_=='ADJ':
                print(token.text, token.pos_,token.dep_,[child])
```

# Get all lematized nouns adjectives and negations       

```python
file_review = "mobile_phones_review.csv"
filename = "lemma_opinion_reviewa.csv"
headers = "id,review,rating"

f = open(filename,"w")
f.write(headers +"\n")
with open(file_review) as csv_file:
    csv_reader = csv.reader(csv_file, delimiter='\t')
    line_count = 0
    for row in csv_reader:
        if line_count == 0:
            line_count +=1
            continue;
        else:
            review = nlp(row[3])
            stropinion = ''
            strnegation=''
            for token in review :
                if token.pos_ =="ADJ":
                    for child in token.children:
                        if child.pos_=='NOUN' and token.lemma_!='στην' and token.lemma_!='στις' and token.lemma_!='...':
                            if nlp.vocab[child.text].is_stop== False:
                                stropinion= stropinion +' ' +token.lemma_  +' ' +child.lemma_
                               
                        elif str(child.lemma_) in neg_list:
                            strnegation = strnegation +'(' +child.lemma_ +')'
                            stropinion = stropinion +' ' +strnegation
                
                elif token.pos_ =="NOUN":
                    for child in token.children:
                        if child.pos_=='ADJ' and child.lemma_!='στην' and child.lemma_!='στις' and child.lemma_!='...':
                            if nlp.vocab[child.text].is_stop== False:
                                stropinion= stropinion +' ' +child.lemma_ +' ' +token.lemma_ 
                        elif str(child.lemma_) in neg_list:
                            strnegation = strnegation +'(' +child.lemma_ +')'
                            stropinion = stropinion +' ' +strnegation 
        
        if int(row[2]) <3:
            rating = 'NEG'
        elif int(row[2]) ==3:
            rating = 'NEU'
        else:
            rating = 'POS'
        f.write(str(line_count) + "," +stropinion + "," +rating +"\n")
        print(str(line_count) +stropinion)
        line_count +=1
f.close()
```
